## Data Cleaning
### Society

```python
# society
flat['society'].value_counts()
```

```python
flat['society'].value_counts().shape
```

```python
import re
```

```python
flat['society'] = flat['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★','',str(name)).strip()).str.lower()
```

### Price

```python
flat['price'].value_counts()
```

```python
flat = flat[flat['price']!='Price on Request']
flat.head()
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
flat['price'] = flat['price'].str.split(' ').apply(treat_price)
```

```python
flat.sample(5)
```

### price_per_sqft

```python
flat['price_per_sqft'].value_counts()
```

```python
flat['price_per_sqft'] = flat['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```

### bedRoom

```python
flat.bedRoom.value_counts() 
```

```python
flat[flat.bedRoom.isnull()]
```

```python
flat = flat[~flat['bedRoom'].isnull()]
```

```python
flat['bedRoom'] = flat['bedRoom'].str.split(' ').str.get(0).astype('int')
```

###  bathroom

```python
flat['bathroom'].value_counts()
```

```python
flat['bathroom'].isna().sum()
```

```python
flat['bathroom'] = flat['bathroom'].str.split(' ').str.get(0).astype('int')
```

### balcony

```python
flat['balcony'].value_counts()
```

```python
flat['balcony'].isna().sum()
```

```python
flat['balcony'] = flat['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

### additionalRoom

```python
flat['additionalRoom'].value_counts()
```

```python
flat['additionalRoom'].value_counts().shape
```

```python
flat['additionalRoom'].isna().sum()
```

```python
flat['additionalRoom'].fillna('not available',inplace = True)
```

```python
flat['additionalRoom'] = flat['additionalRoom'].str.lower()
```

### floorNum

```python
flat['floorNum'].value_counts()
```

```python
flat[flat['floorNum'].isna()]
```

```python
flat['floorNum'] = flat['floorNum'].str.split(' ').str.get(0).replace('Ground','0').str.replace('Basement','-1').str.replace('Lower','0').str.extract(r'(\d+)')
```

### facing

```python
flat['facing'].value_counts()
```

```python
flat['facing'].isna().sum()
```

```python
flat['facing'].fillna('NA',inplace =True)
```
